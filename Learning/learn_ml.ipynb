```python
import pandas as pd
import os
```

```python
melbourne_file_path = '/Users/twinklemittal/Desktop/AI/Machine_learning/machine_learning_algorithms/original_data/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path)
print(len(melbourne_data))
melbourne_data.describe()
```

```python
# dropna drop missing value
# data preprocessing
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
y = melbourne_data.Price
y
```

```python
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
X = melbourne_data[melbourne_features]
X
```

```python
# Building the model
from sklearn.tree import DecisionTreeRegressor

# random_state = 1, it menas when we will run this notebook again and again then it will give the same result for each time.
melbourne_model = DecisionTreeRegressor(random_state=1)

# fit model
melbourne_model.fit(X, y)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predicted_home_prices = melbourne_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    melbourne_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    melbourne_model.fit(train_X, train_y)
    prediction = melbourne_model.predict(val_X)
    mae = mean_absolute_error(val_y, prediction)
    return mae
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    print(get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y))
    
```
