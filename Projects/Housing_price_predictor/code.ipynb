```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
```

```python
home_data = pd.read_csv("/Users/twinklemittal/Desktop/AI/Machine_learning/machine_learning_algorithms/Projects/Housing_price_predictor/Dataset/train.csv")
```

```python
y = home_data.SalePrice
```

```python
home_data.columns
```

```python
features = ['LotArea', 'BedroomAbvGr','1stFlrSF', '2ndFlrSF', 'YearBuilt', 'TotRmsAbvGrd']
X = home_data[features]
```

```python
X
```

```python
home_pred_model = DecisionTreeRegressor(random_state=0)
home_pred_model.fit(X, y)
home_pred_model.predict(X.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
home_pred_model = RandomForestRegressor(random_state=1)
```

```python
home_pred_model.fit(train_X, train_y)
predicted = home_pred_model.predict(val_X)
mean_absolute_error(val_y, predicted)
```

```python
# Train complete data
complete_home_pred_model = RandomForestRegressor(random_state=1)
complete_home_pred_model.fit(X, y)
```

```python
test_data = pd.read_csv("/Users/twinklemittal//Desktop/AI/Machine_learning/machine_learning_algorithms/Projects/Housing_price_predictor/Dataset/test.csv")
```

```python
test_data_X = test_data[features]
test_data.columns
```

```python
new_complete_prediction = complete_home_pred_model.predict(test_data_X)
```

```python
new_complete_prediction
```

```python
output= pd.DataFrame({'id': test_data.Id, 'SalePrice': new_complete_prediction})
```
