```python
import numpy as np
import matplotlib.pyplot as plt

# Number of coin flips per experiment
n = 10

# Number of times we repeat the experiment (higher = smoother curve)
num_experiments = 10000

# Simulating the experiment
results = []
for _ in range(num_experiments):
    flips = np.random.choice([0,1], size=n, p=[0.5, 0.5]) # 0 = Tails, 1 = Heads
    num_heads = np.sum(flips)
    results.append(num_heads)

print("Simulated Results:")
print(results)

# Count occurrences of each possible number of heads (0 to 10)
values, counts = np.unique(results, return_counts=True)

# Convert counts to probabilities
probabilities = counts / num_experiments

# Plotting
plt.figure(figsize=(8,5))
plt.bar(values, probabilities, color='lightcoral', edgecolor='black', alpha=0.7, label="Simulated Probabilities")
plt.plot(values, probabilities, 'o-', color='blue', label="Smooth Curve") # Connecting dots for smoothness

# Labels and Title
plt.xlabel("Number of Heads in 10 coin Flips")
plt.ylabel("Probability")
plt.title("Simulated Probability Distribution of Getting Heads in 10 Coin Flips")
plt.xticks(np.arange(0, n+1, 1))
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show Plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of dice rolls per experiment
num_experiments = 10000

# Simulating the experiment
results = []
for _ in range(num_experiments):
    die1 = np.random.randint(1,7)   # Roll first die (1 to 6)
    die2 = np.random.randint(1,7)  # Roll second die (1 to 6)
    total = die1 + die2 # Sum of both dice
    results.append(total)

print("Simulated Results:")
print(results)

# Count occurences of each possible sum (2 to 12)
values, counts = np.unique(results, return_counts=True)

# Convert counts to probabilities
probabilities = counts / num_experiments

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(values, probabilities, color='orange', edgecolor='black', alpha=0.7, label="Simulated Probabilities")
plt.plot(values, probabilities, 'o-', color='blue', label="Smooth Curve")  # Connecting dots for smoothness

# Labels and Title
plt.xlabel("Sum of Two Dice Rolls")
plt.ylabel("Probability")
plt.title("Simulated Probability Distribution of Sum of Two Dice Rolls")
plt.xticks(np.arange(2, 13, 1))
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show Plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sum of two Dice (possible values: 2 to 12)
values = np.arange(2, 13)

# Theoretical probabilities based on counting combinations
probabilities = np.array([
    1/36,  # Sum = 2 (only 1+1)
    2/36,  # Sum = 3 (1+2, 2+1)
    3/36,  # Sum = 4 (1+3, 2+2, 3+1)
    4/36,  # Sum = 5 (1+4, 2+3, 3+2, 4+1)
    5/36,  # Sum = 6 (1+5, 2+4, 3+3, 4+2, 5+1)
    6/36,  # Sum = 7 (most common: 1+6, 2+5, 3+4, etc.)
    5/36,  # Sum = 8
    4/36,  # Sum = 9
    3/36,  # Sum = 10
    2/36,  # Sum = 11
    1/36   # Sum = 12 (only 6+6)
])

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(values, probabilities, color='orange', edgecolor='black', alpha=0.7, label="Theoretical Probabilities")
plt.plot(values, probabilities, 'o-', color='blue', label="Smooth Curve")  # Connecting dots for smoothness

# Labels and Title
plt.xlabel("Sum of Two Dice Rolls")
plt.ylabel("Probability")
plt.title("Theoretical Probability Distribution of Sum of Two Dice Rolls")
plt.xticks(values)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show Plot
plt.show()
```
